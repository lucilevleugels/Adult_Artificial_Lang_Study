```python
def create_repetitions(repetition, df):
    
    row_copy = df.iloc[repetition-1,:]
    row_copy = pd.DataFrame(row_copy).T

    before_r = df.iloc[:repetition, :]
    after_r = df.iloc[repetition:, :]

    test = pd.concat([before_r, row_copy, after_r],axis=0)
        
    return test
```
